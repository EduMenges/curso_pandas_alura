```python
import pandas as pd
dados = pd.read_csv("dados/aluguel_residencial.csv", sep=";")
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
```

```python
somente_bairros = dados[dados["Bairro"].isin(bairros)]
somente_bairros
```

```python
agrupados = dados.groupby("Bairro")
agrupados.groups
```

```python
# Nós temos um dicionário, com o bairro (chave) e os dados (valor).
# Os valores são dataframes. Exemplo, pegando a média de aluguel dentro dos bairros:

for bairro, dados in agrupados:
    print(f"{bairro}: {round(dados['Valor'].mean(), 2)}")
```

```python
# Usando uma sintaxe mais concisa:

agrupados["Valor"].mean().round(2)
```

```python
# Desta vez, uma série é retornada. Se especificar mais de uma coluna, um DF é retornado

agrupados[["Valor", "Condominio"]].mean().round(2)
```
