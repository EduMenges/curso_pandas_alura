## Seleções e Frequências

```python
import pandas as pd
dados = pd.read_csv("dados/somente_residenciais.csv", sep=";")
```

### Selecione somente os imóveis classificados com tipo 'Apartamento'.

```python
selecao1 = dados[dados["Tipo"] == "Apartamento"]
selecao1.head()
```

### Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' ou 'Casa de Vila'.

```python
selecao2 = dados[dados['Tipo'].isin(['Casa', 'Casa de Condomínio', 'Casa de Vila'])]
selecao2.head()
```

### Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

```python
selecao3 = dados[(dados["Area"] <= 100) & (dados["Area"] >= 60)]
selecao3.head()
```

### Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

```python
selecao4 = dados[(dados["Valor"] < 2000) & (dados["Quartos"] >= 4)]
selecao4.head()
```
