# Relatório de Análise III

## Imóveis Residenciais

```python
import pandas as pd
dados = pd.read_csv(filepath_or_buffer="dados/aluguel.csv", sep=";")
```

```python
dados.tail()
```

## Pegando apenas os tipos residenciais

```python
tipos_totais = list(dados["Tipo"].drop_duplicates())
tipos_totais
```

```python
tipos_residenciais = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Flat',
 'Casa de Vila',
 'Loft',
 'Loteamento/Condomínio',
 'Studio']
```

## Pegando somente os registros residenciais

```python
somente_residenciais = dados[dados["Tipo"].isin(tipos_residenciais)]
somente_residenciais
```

```python
somente_residenciais["Tipo"].drop_duplicates()
```

```python
somente_residenciais.index = range(somente_residenciais.shape[0])
somente_residenciais
```

# Exportando a Base de Dados

## Usando csv

```python
somente_residenciais.to_csv(path_or_buf="dados/somente_residenciais.csv", sep=";", index=False)
```
