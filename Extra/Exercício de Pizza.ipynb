```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 7))

dados = pd.read_csv("../dados/aluguel_amostra.csv", sep=";")
```

```python
# Total de imóveis por Tipo

por_tipo = dados.groupby(["Tipo"])["Valor"]
por_tipo.count()
```

```python
# Total de imóveis por Tipo Agregado

por_agregado = dados.groupby(["Tipo Agregado"])["Valor"]
por_agregado.count()
```

```python
area = plt.figure()

g1 = area.add_subplot(1,2,1)
g2 = area.add_subplot(1,2,2)

valores = por_agregado.count().values
label = por_agregado.count().index
g1.pie(valores, labels=label, autopct="%1.1f%%", explode=(.1,.1))

valores = por_tipo.count().values
label = por_tipo.count().index
g2.pie(valores, labels=label, autopct="%1.1f%%", explode=(.1,.1,.1,.1,.1))
```
